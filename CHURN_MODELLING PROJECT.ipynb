```python
import pandas as pd              # For loading and manipulating the CSV dataset
import numpy as np               # For numerical operations, e.g., creating mesh grid
from sklearn.model_selection import train_test_split  # For splitting data into training and test sets
from sklearn.ensemble import AdaBoostClassifier,GradientBoostingClassifier # For implementing the AdaBoost algorithm
import lightgbm as lgb
from sklearn.tree import DecisionTreeClassifier       # For creating decision stumps (weak learners)
from sklearn.metrics import accuracy_score, classification_report,confusion_matrix  # For evaluating model performance
from sklearn.preprocessing import StandardScaler , LabelEncoder
```

```python
df = pd.read_csv(r"C:\Users\user\Music\Machine learning\SUPERVISED LEARNING\samples\SAMPLE DATASETS\Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.isna().sum() # FINDING NULL VALUE
```

```python
Label = LabelEncoder() ## converting categorical text data into numberes
df["Geography"] = Label.fit_transform(df["Geography"])
```

```python
Label = LabelEncoder()
df["Gender"] = Label.fit_transform(df["Gender"])
```

```python
df.sample(10)
```

```python
# Selecting input(x) and output(y) columns
x = df[["CreditScore","Geography","Gender","Age","Tenure","Balance","NumOfProducts","HasCrCard","IsActiveMember","EstimatedSalary"]]
y = df["Exited"]  
```

```python
x
```

```python
X_train , X_test , y_train , y_test = train_test_split(x,y , test_size=0.2,random_state=42)
scaler = StandardScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

# ADA BOOST

```python
weak_learner = DecisionTreeClassifier(max_depth=1) # using DecisionTreeClassifier
ada = AdaBoostClassifier(estimator=weak_learner, n_estimators=50, learning_rate=0.01, random_state=42) # using ada boost algrothim
```

```python
ada.fit(X_train_scaled,y_train)
```

```python
y_pred = ada.predict(X_test_scaled) # predicting on test value
y_pred
```

```python
y_test
```

```python
accuracy_score(y_test,y_pred) # checking accuracy on test data
```

```python
print(classification_report(y_test,y_pred))  # classifcation report on test data
```

```python
confusion_matrix(y_test,y_pred)
```

```python
y_pred_t = ada.predict(X_train_scaled) # predicting of train data
y_pred_t
```

```python
y_train
```

```python
accuracy_score(y_train,y_pred_t) # accuracy of train data
```

```python
print(classification_report(y_train,y_pred_t))  # classifaction of train data
```
